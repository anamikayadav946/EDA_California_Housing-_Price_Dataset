## **<font color = 'blue'>6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.</font>**

```python
def median_cal(data):
  data=np.sort(data.values)
  n = len(data)
  if n%2 == 1:
    sum = data[n//2]
  else:
    mid1 = data[n//2]
    mid2 = data[(n-1)//2]
    sum = (mid1 + mid2)/2

  return sum

mv = median_cal(housing['median_income'])
mv
```

```python
mv == housing['median_income'].median()
```

## **<font color = 'blue'>10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large</font>**

```python
def get_bedroom_size_category(row):
  total_bedrooms = row['total_bedrooms']
  if total_bedrooms <= 10:
    return 'small'
  elif total_bedrooms < 1000:
    return 'medium'
  else:
    return 'large'

# Apply the function to each row and create a new column
housing['total_bedroom_size'] = housing.apply(get_bedroom_size_category, axis=1)
housing.head(2)
```
